```python
import numpy
import pandas
import matplotlib.pyplot

I = [i + 1 for i in range(0,10)]
II = [(i + 1) ** 2 for i in range(0,10)]

df = pandas.DataFrame({'Numbers':I, 'Square':II})
print(df)


III = [x for x in I if x % 2 == 0]
IV = [x**2 for x in III]

dff  = pandas.DataFrame({'Numbers':III, 'Square':IV})
dff = dff.drop('Numbers', axis='columns')
print(dff)
```
